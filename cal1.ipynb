```python
#importing and converting .dat files

#df = pd.read_csv('u11_49T170U32_84.dat', delimiter='\t')

#data = np.genfromtxt('/Users/larsbaltensperger/Desktop/Physikpraktikum3/Franck_Hertz/plot1/u11_49T170U32_84.dat',dtype=None,delimiter=',')

data1 = np.genfromtxt('/Users/larsbaltensperger/Desktop/Physikpraktikum3/Franck_Hertz/plot1/u11_50T171u32_70.txt')
data2 = np.genfromtxt('/Users/larsbaltensperger/Desktop/Physikpraktikum3/Franck_Hertz/plot1/u11_50T171u33_27.txt')
data3 = np.genfromtxt('/Users/larsbaltensperger/Desktop/Physikpraktikum3/Franck_Hertz/plot1/u11_50T171u33_90.txt')

#print(data.shape)
#print(data)

#print(data[:,0])

u2_1 = data1[:,1]*10
uA_1 = data1[:,2]

u2_2 = data2[:,1]*10
uA_2 = data2[:,2]

u2_3 = data3[:,1]*10
uA_3 = data3[:,2]
```
